## 組合

```python
import csv
import numpy as np
import re

file_name = "input.fasta.map" #在這裡輸入要整理的檔案名稱
output_file_name = "fasta_data_clean.csv" #在這裡輸入整理完的檔案名稱

with open(file_name) as f:
    content = f.read()
std_try = "(>+[A-Za-z0-9_|-]+)+\n+(#+[A-Za-z0-9, ]+)+\n+([a-z0-9 ,-/(\n/)]+)"
d = re.findall(std_try,content)

with open(output_file_name, 'w', newline='') as csvfile:
    for w in d:
        e = w[2].replace(",","").split()
        table = np.array([
                      [w[0]],
                      [w[1]],
                      [" "]
                      ], dtype=object)
        for i in range(0,len(e),3):
            column = np.array([
                [e[i]],
                [e[i+1]],
                [e[i+2]]
                ])
            table = np.column_stack((table,column))    
        writer = csv.writer(csvfile)
        writer.writerows(table)   
```
