```python
import pandas as pd
```

```python
taiwan_data = pd.read_csv("data/tai.csv",encoding = "utf-8")
taiwan_data.head()
```

```python
world_data = pd.read_excel("data/world.xlsx",encoding = "utf-8")
world_data.head()
```

```python
in_world = []
count = 0
here = 0
for i in taiwan_data["genus"]:
    for j in world_data["Name_La"]:
        if i == j:
            in_world.append("1")
            count += 1
            here += 1
            break
    else:
        in_world.append("0")
        count += 1
        print(i)
print("taiwan_total : ",count)
print("taiwan與world有匹配的數量 : ",here)
taiwan_data["in_world"] = in_world
```

```python
taiwan_data.to_csv("tai_test.csv",encoding = "utf-8-sig")
```

```python
in_taiwan = []
count = 0
here = 0
for i in world_data["Name_La"]:
    for j in taiwan_data["genus"]:
        if i == j:
            in_taiwan.append("1")
            count += 1
            here += 1
            #print(i)
            break
    else:
        in_taiwan.append("0")
        print(i)
        count += 1
print("world_total : ",count)
print("world與taiwan有匹配的數量 : ",here)
world_data["in_Taiwan"] = in_taiwan
```
